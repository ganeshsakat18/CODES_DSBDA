```python
import pandas as pd
import numpy as np
```

```python
missing_values=['?',np.nan,'[]']
```

```python
df1=pd.read_csv('movies_metadata.csv', na_values=missing_values)
df2=pd.read_csv('credits.csv', na_values=missing_values)
df3=pd.read_csv('links_small.csv', na_values=missing_values)
df4=pd.read_csv('keywords.csv', na_values=missing_values)
df5=pd.read_csv('ratings_small.csv', na_values=missing_values)
df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
```

```python
df.info()
```

```python
df
```

```python
df.isnull().sum()
```

```python
def clean(data):
    for col in data:
        df[col].fillna(method='ffill', inplace=True)
        
clean(df)        
```

```python
df = df.dropna(subset=(['cast']))
```
