```python
soup=bs(page.content,'html.parser')
```

```python
print(soup.prettify())
names = soup.find_all('p', class_='_2NsDsF AwS1CA') 
```

```python
names
```

```python
#Extracting customer names
cust_name=[]
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
```

```python
cust_name
```

```python
reviews = soup.find_all('div', class_= 'XQDdHH Ga3i8K') 
```

```python
reviews
```

```python
#Extracting ratings names
cust_rrating=[]
for i in range(0,len(reviews)):
    cust_rrating.append(reviews[i].get_text())
```

```python
cust_rrating
```

```python
comments= soup.find_all('div', class_='ZmyHeo') 
```

```python
comments
```

```python
#Extracting ratings names
cust_comments=[]
for i in range(0,len(comments)):
    cust_comments.append(comments[i].get_text())
```

```python
df = pd.DataFrame() 
df['Customer Name']=cust_name 
df['Ratings']=cust_rrating
df['Reviews']=cust_comments
```
